## Getting all entries containing a particular element

```python
a = MPRester("EDFQtMk1T0w7CbJu")
criteria = {"elements":{"$all": ["Ti"]}}
properties = [u"pretty_formula", u"formula", u"density", u"band_gap", u"volume", u"spacegroup"]
titanium = a.mpquery(criteria, properties)
criteria = {"elements":{"$all":["Zr"]}}
zirconium = a.mpquery(criteria, properties)
```

We want to compare compounds were titanium has been replaced with zirconium.  So, we can construct the forumula and search through the other list for a match.

```python
def find_match(formula, thelist):
    for item in thelist:
        if item['formula'] == formula:
            yield item
```

```python
for item in find_match(formula={u'Ti':1.0, u'O':2.0}, thelist=titanium):
    print item['spacegroup'], item['band_gap']
```

Okay, so there are multiple entries for at least some of the formulas. Some of those are different spatial configurations, but not all. I'm not sure where the distinct entries with the same spatial group information come from; it seems to me that they ought to agree better than they do. Of course, these values are reported without any uncertainties, so it's not clear if the disagreement is real (though it looks significant enough that it would have to be).

Let's see how many distinct formulas we have in each list.

```python
def count_formulas(compoundlist):
    distincts = set([item['pretty_formula'] for item in compoundlist])
    return len(distincts)
```

```python
count_formulas(titanium)
```

```python
count_formulas(zirconium)
```

Still have more than twice as many for Ti as for Zr. Of these, how many match?

```python
ti_formulas = set([item['pretty_formula'] for item in titanium])
zr_formulas = set([item['pretty_formula'] for item in zirconium])
converted = set()
for formula in ti_formulas:
    converted.add(formula.replace("Ti", "Zr"))
common = converted & zr_formulas
len(common)
```

Let's compare volumes.  We have a problem, which is that some of these pretty formulas correspond to multiple entries in the original lists, but they may very well not be the same number. Even for those that have the same number of entries, it's important to match up the matching ones. And, since I'm not 100% sure what makes the separate entries distinct, I don't really know how to do that well.

So, reasonable sounding question at this point: how many of these are multiples, and how many multiples do they have?

```python
singles = 0
matches = 0
differs = 0
singleset = set()
for formula in common:
    zrformula = formula
    tiformula = formula.replace("Zr", "Ti")
    ticount = 0
    zrcount = 0
    for tiitem in titanium:
        if tiitem['pretty_formula']==tiformula:
            ticount += 1
    for zritem in zirconium:
        if zritem['pretty_formula']==zrformula:
            zrcount += 1
    #print formula, ticount, zrcount
    if zrcount == ticount:
        if zrcount == 1:
            singles += 1
            singleset.add(formula)
        else:
            matches += 1
    else:
        differs += 1
print "%d compunds differ; %d singles (%d other)" %(differs, singles, matches)
```

Variability from sources other than the single replacement will confuse what I'm trying to look at, so I'll only keep those that have a single entry for each element.  This brings the number down to 120 (from a starting number of 2560 for titanium), so I've dramatically reduced the number of compounds I'm looking at.

Comparisons are getting awkward. I'd really like to have these things in a dataframe so I can manipulate them more easily. The majority of that looping and conditional nonsense in the previous cell would have gone away if I had moved into pandas in the first place. Let me do that now.

```python
def get_two_elements(eone, etwo):
    a = MPRester("EDFQtMk1T0w7CbJu")
    properties = [u"pretty_formula", u"formula", u"density", u"band_gap", u"volume", u"spacegroup", u"energy"]
    
    # element one
    criteria = {"elements":{"$all": [eone]}}
    eonelist = a.mpquery(criteria, properties)
    eonedf = pd.DataFrame(eonelist)

    # element two
    criteria = {"elements":{"$all":[etwo]}}
    etwolist = a.mpquery(criteria, properties)
    etwodf = pd.DataFrame(etwolist)
    
    return eonedf, etwodf
```

```python
titanium['base_formula'] = titanium['pretty_formula'].map(make_generic)

def get_multi_elements(elist):
    a = MPRester("EDFQtMk1T0w7CbJu")
    properties = [u"pretty_formula", u"formula", u"density", u"band_gap", u"volume", u"spacegroup", u"energy"]

    resultslist = []
    for element in elist:
        def make_generic(formula):
            return formula.replace(element, "??")
     
        criteria = {"elements":{"$all":[element]}}
        singlelist = a.mpquery(criteria, properties)
        eldf = pd.DataFrame(singlelist)
        eldf['element'] = element
        eldf['base_formula'] = eldf['pretty_formula'].map(make_generic)
        eldf['counts'] = eldf.groupby('pretty_formula')['energy'].transform('count')
        resultslist.append(eldf)
    
    return resultslist
```

```python
def biggest_groups(dflist, printvals=False):

    # build the set of the compounds with the most entries
    baseset = set()
    for thedf in dflist:
        maxidx = thedf['counts'].idxmax()
        baseset.add(thedf['base_formula'][maxidx])
    # for each of the base formulas, extract the appropriate name and count from each dataframe.
    if printvals:
        for baseformula in baseset:
            print "For base formula " + baseformula + " : "
            for thedf in dflist:
                try:
                    theformula = thedf[thedf['base_formula']==baseformula]['pretty_formula'].iloc[0]
                    thecount = thedf[thedf['base_formula']==baseformula]['counts'].iloc[0]
                    print "  %s: %d entries  " % (theformula, thecount)
                except(IndexError):
                    print "not found"
    return baseset
```

```python
def generate_box_plots(dflist, baseset, variable):
    for baseformula in baseset:
        subdfs = []
        for thedf in dflist:
            subdfs.append(thedf[thedf['base_formula']==baseformula])
        subbed = pd.concat(subdfs)
        subbed.boxplot(column=variable, by='pretty_formula')
```
